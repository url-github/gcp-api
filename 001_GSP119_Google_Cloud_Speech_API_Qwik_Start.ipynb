`curl`: Polecenie do przesyłania żądań HTTP z wiersza poleceń.

`-s`: Opcja, która oznacza tryb "ciszy" (silent), co oznacza, że curl nie będzie wyświetlać postępu ani informacji o transferze danych.

`-X POST`: Opcja określająca metodę żądania HTTP, w tym przypadku POST.

`-H "Content-Type: application/json"`: Opcja określająca nagłówek żądania HTTP, który mówi serwerowi, że dane są przesyłane w formacie JSON.

`--data-binary @request.json`: Opcja przekazująca dane do wysłania. W tym przypadku dane są pobierane z pliku request.json i przesyłane binarnie.

`"https://speech.googleapis.com/v1/speech:recognize?key=${API_KEY}"`: Adres URL, do którego jest wysyłane żądanie.

```python
curl -s -X POST -H "Content-Type: application/json" --data-binary @request.json \
"https://speech.googleapis.com/v1/speech:recognize?key=${API_KEY}"
```
